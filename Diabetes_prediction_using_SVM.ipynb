#Diabetes prediction using SVM

```python

#import the necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes_data=pd.read_csv('/content/diabetes.csv')
print(diabetes_data)
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
x=diabetes_data.drop(columns='Outcome',axis=1)
y=diabetes_data['Outcome']
```

```python
print(x)
print(y)
```

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
print(standardized_data)
```

```python
[x_train,x_test,y_train,y_test]=train_test_split(x,y,train_size=0.90,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Load/import model
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
#Accuracy score on training data
x_train_prediction=classifier.predict(x_train)
x_train_data_accuracy=accuracy_score(x_train_prediction,y_train)
print(x_train_data_accuracy)
```

```python
#creating confusion matrix,which compares the y_test and y_pre
cm = confusion_matrix(x_train_prediction, y_train)
cm
```

```python
#Generate classification report
print(classification_report(x_train_prediction, y_train))
```

```python
# SVM model performance on test data
x_test_prediction=classifier.predict(x_test)
x_test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print(x_test_data_accuracy)
```

```python
unseen_data=(11,143,94,33,146,36.6,0.254,51)

# changing the input_data to numpy array
unseen_data_as_numpy_array = np.asarray(unseen_data)

# reshape the array as we are predicting for one instance
unseen_data_reshaped = unseen_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(unseen_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
